```python
#A python list of lists representing a sudoku puzzle, where 0 represents an empty cell.

sudoku_list = [[0, 0, 4, 3, 0, 0, 2, 0, 9],
 [0, 0, 5, 0, 0, 9, 0, 0, 1],
 [0, 7, 0, 0, 6, 0, 0, 4, 3],
 [0, 0, 6, 0, 0, 2, 0, 8, 7],
 [1, 9, 0, 0, 0, 7, 4, 0, 0],
 [0, 5, 0, 0, 8, 3, 0, 0, 0],
 [6, 0, 0, 0, 0, 0, 1, 0, 5],
 [0, 0, 3, 5, 0, 8, 6, 9, 0],
 [0, 4, 2, 9, 1, 0, 3, 0, 0]]


```

```python
import numpy as np

# Convert the list of lists to a numpy array called sudoku_array
sudoku_array = np.array(sudoku_list) # Convert the list of lists to a numpy array called sudoku_array
print(sudoku_array) # Print the sudoku_array to check it was created correctly
```

```python
print(type(sudoku_array)) # Check the type of sudoku_array
```
