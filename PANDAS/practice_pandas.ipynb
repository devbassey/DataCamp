# WHAT IS PANDAS

Pandas is a Python library for working with structured data. Think tables. Like Excel sheets. Like SQL tables. But inside Python, and way more flexible.



```python
import pandas as pd
```

```python
data = {
    'name': ['utibe', 'bassey','inyang', 'godwin'],
    'age': [30, 20, 45, 50],
    'location': ['lagos', 'uyo', 'abuja', 'rivers']
}

df = pd.DataFrame(data)
df

display(data)
```

DataFrame, is a full table. Rows and columns. This is what you will live in 90% of the time.

- df.head() - shows the first few rows
- df.shape() - tells your rows and column
- df.info() - tells you the real story: data types, missing values, memory. etc.

Because everything you’ll ever do… cleaning, analyzing, visualizing, modeling… starts with pulling out the right piece of the table.

In Pandas, Rows have "index" and Columns have "labels".



# Talking to Columns

To get a single column, call the individual column name inside a square bracket;

e.g: df['name']

This returns a Series. One column. With an index.

```python
# Get the 'age' column
df['age']
```

```python
# Get the name column
df['name']
```

```python
# Get the location column
df['location']
```

You can also use the dot notation to access column values;

df.name - for instance

```python
# Get the name column using dot notation
df.name
```

```python
# Get multiple columns: name and location
df[['name', 'location']]
```

Notice the brackets
- one bracket = series
- two brackets = DataFrame
